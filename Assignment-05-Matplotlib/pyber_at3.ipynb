```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
merged_data = pd.merge(ride_data, city_data, on="city")
# Display the data table for preview
merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
data_grouped_df = merged_data.groupby("city")
data_grouped_df.head()
```

```python
cities = data_grouped_df["city"].unique()
types = data_grouped_df["type"].unique()
fare_average = data_grouped_df["fare"].mean()
total_rides = data_grouped_df["city"].count()
driver_count = data_grouped_df["driver_count"].mean()
fare_total = data_grouped_df["fare"].sum()
```

```python
rides_data_df = pd.DataFrame({
                            "Average Fare" : fare_average,
                            "Total Fare" : fare_total,
                            "Total Rides" : total_rides,
                            "Driver Count" : driver_count,
                            "Type" : types})
rides_data_df
```

```python
urban_rides = rides_data_df.loc[rides_data_df["Type"]=="Urban"]
suburban_rides = rides_data_df.loc[rides_data_df["Type"]=="Suburban"]
rural_rides = rides_data_df.loc[rides_data_df["Type"]=="Rural"] 
```

```python
# Obtain the x and y coordinates for each of the three city types
# x_axis = count of each city
urban_x_axis = urban_rides["Total Rides"]
suburban_x_axis = suburban_rides["Total Rides"]
rural_x_axis = rural_rides["Total Rides"]
# y_axis = average of fare for each city
urban_y_axis = urban_rides["Average Fare"]
suburban_y_axis = suburban_rides["Average Fare"]
rural_y_axis = rural_rides["Average Fare"]
# s = total_driver_count
urban_size = urban_rides["Driver Count"]
suburban_size = suburban_rides["Driver Count"]
rural_size = rural_rides["Driver Count"]

# Build the scatter plots for each city types
plt.figure(figsize=(12,5))
urban_scatter = plt.scatter(urban_x_axis, urban_y_axis, s=urban_size*10, alpha=0.75, edgecolors="black", c="coral")
suburban_scatter = plt.scatter(suburban_x_axis, suburban_y_axis, s=suburban_size*10, alpha=0.75, edgecolors="black", c="lightblue")
rural_scatter = plt.scatter(rural_x_axis, rural_y_axis, s=rural_size*10, alpha=0.75, edgecolors="black", c="gold")

plt.grid()


# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data")
# Create a legend
plt.legend(handles=[urban_scatter, suburban_scatter, rural_scatter],
           labels=["Urban", "Suburban", "Rural"], 
                   loc="best", markerscale=1, title="City Types", labelspacing=1)
# Incorporate a text label regarding circle size
plt.text(42, 35, "Note:\nCircle size\ncorrelates\nwith driver\ncount per\ncity.")

# Save Figure
plt.savefig("../Images/Pyber_BubblePlot.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fares = merged_data["fare"].sum()
urban_fares = merged_data.loc[merged_data["type"]=="Urban","fare"].sum()
suburban_fares = merged_data.loc[merged_data["type"] == "Suburban", "fare"].sum()
rural_fares = merged_data.loc[merged_data["type"] == "Rural", "fare"].sum()

urban_fares_percent = round(urban_fares/total_fares,3)*100
suburban_fares_percent = round(suburban_fares/total_fares,3)*100
rural_fares_percent = round(rural_fares/total_fares,3)*100

overall_fares = [urban_fares_percent,  rural_fares_percent,suburban_fares_percent]
explode = [0.1, 0 , 0]
types = ["Urban",  "Rural","Suburban"]
colors = ["lightcoral",  "gold","lightskyblue"]

# Build Pie Chart
plt.pie(overall_fares, explode=explode, labels = types, colors= colors, autopct="%1.1f%%", shadow=True,startangle=300)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("../Images/Pyber_TotalFares.png")
plt.axis("equal")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_ride_count = rides_data_df["Total Rides"].sum()
urban_ride_count = rides_data_df.loc[rides_data_df["Type"] == "Urban" , "Total Rides"].sum()
suburban_ride_count = rides_data_df.loc[rides_data_df["Type"] == "Suburban" , "Total Rides"].sum()
rural_ride_count = rides_data_df.loc[rides_data_df["Type"] == "Rural" , "Total Rides"].sum()

urban_ride_percent = round(urban_ride_count/total_ride_count, 3)*100
suburban_ride_percent = round(suburban_ride_count/total_ride_count, 3)*100
rural_ride_percent = round(rural_ride_count/total_ride_count, 3)*100
# Build Pie Charts
overall_rides = [urban_ride_percent, rural_ride_percent, suburban_ride_percent]
explode = [0.1, 0 , 0]
types = ["Urban",  "Rural","Suburban"]
colors = ["lightcoral",  "gold","lightskyblue"]

plt.pie(overall_rides, colors=colors, explode=explode, labels=types, autopct="%1.1f%%", startangle=250, shadow=True)
plt.title("% of Total Rides by City Type")
plt.axis("equal")
# Save Figure
plt.savefig("../Images/Pyber_TotalRides.png")
plt.show()
```
